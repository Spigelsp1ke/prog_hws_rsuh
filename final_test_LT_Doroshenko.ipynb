#### Задача 3.

Вводится список. Если список упорядочен по возрастанию, выведите "по возрастанию", если по убыванию - то "по убыванию", а если не упорядочен - то "не упорядочен".

```python
def checker(lst):
    if sorted(lst) == lst:
        return ('по возрастанию')
    elif sorted(lst, reverse=True) == lst:
        return ('по убыванию')
    else:
        return ('не упорядочен')
    
checker([1, 2, 4, 3])
```

#### Задача 5.

Программе на вход подается число в [римской записи](https://ru.wikipedia.org/wiki/%D0%A0%D0%B8%D0%BC%D1%81%D0%BA%D0%B8%D0%B5_%D1%86%D0%B8%D1%84%D1%80%D1%8B), то есть, строка содержит заглавные латинские буквы: I, V, X, C, D, M, причем IV - это 4, XC - это 90, CM - это 900. Программа должна переводить это число в арабское.

```python
roman = { 1000: 'M', 
  900: 'CM', 
  500: 'D', 
  400: 'CD', 
  100: 'C',
  90: 'XC', 
  50: 'L', 
  40: 'XL', 
  10: 'X', 
  9: 'IX', 
  5: 'V', 
  4: 'IV', 
  1: 'I'}
def roman_to_ar(rom):
    ar_num = 0
    for k, v in roman.items():
        while rom.startswith(v):
            ar_num += k
            rom = rom[len(v):]
    return ar_num

roman_to_ar('LXIII')
```

```python
def sequence(x, y):
    count = 0
    while not y == 0:
        y = int(input())
        count += 1
        if y == x:
            return f'Число {x} есть в последовательности под номером {count + 1}'
    return f'Число {x} не встретилось'


x = int(input())
y = int(input())
sequence(x, y)
```

#### Задача 7.

Пользователь вводит текст, в котором встречаются цифры. Напишите функцию, которая будет находить цифры и печатать в столбик.

```python
def numbers(text):
    textlist = text.split()
    num_list = []
    for i in textlist:
        if i.isdigit():
            num_list.append(int(i))
    return num_list


numbers('12345 asdffg 234')
# 12345 asdffg 234 == 12345, 234
```

    <слово1> (<длина1>) - <индекс1>
    <слово2> (<длина2>) - <индекс2>
    ...

Например, если на 7-ом месте встретилось слово 'я', надо вывести: 'я (7) - 1'. Если таких слов нет, выведите на экран "Все слова чётной длины". Также требуется вывести на экран, сколько встретилось слов чётной длины.

```python
def odd_word(string):
    lst_all_words = []
    lst_odd = []
    lst_even = []
    while not string == ' ':
        lst_all_words.append(string)
        if len(string) % 2 == 1:
            lst_odd.append(string)
        else:
            lst_even.append(string)
        string = input()
    if len(lst_odd) == 0:
        print(f"Все слова четной длины. Кол-во слов: {len(lst_even)}")
    else:
        for i in lst_odd:
            for j in lst_all_words:
                if i == j:
                    print(f'{i} ({lst_all_words.index(j)}) - {len(i)}.')
        print(f'Кол-во четных слов: {len(lst_even)}')
           



string = input()
odd_word(string)
```

```python
s1 = input()
s2 = input()
d = dict()
for i in range(len(s1)):
    if s1[i] not in d:
        d[s1[i]] = 1
    else:
        d[s1[i]] += 1
for i in range(len(s2)):
    if s2[i] in d:
        if d[s2[i]] > d[s1[i]]:
            print(f'{s1} - {s2} (не ок)')
        else:
            print(f'{s1} - {s2} (ок)')
    else:
        print(f'{s1} - {s2} (не ок)')
            
        

```
